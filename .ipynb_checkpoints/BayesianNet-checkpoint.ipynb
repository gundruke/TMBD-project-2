```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
df.drop(columns=["id"], inplace=True)
```

```python
df.columns
```

```python
df.describe()
```

```python
df["smoking_status"] = df["smoking_status"].factorize()[0]
df["gender"] = df["gender"].factorize()[0]
df["Residence_type"] = df["Residence_type"].factorize()[0]
df["ever_married"] = df["ever_married"].factorize()[0]
df["work_type"] = df["work_type"].factorize()[0]
```

```python
df
```

```python
df["age"] = (df["age"] > 45).astype(int)
df.rename(columns={"age":"age_above_45"}, inplace=True)

df["avg_glucose_level"] = (df["avg_glucose_level"] > 100).astype(int)
df.rename(columns={"avg_glucose_level":"avg_glucose_level_above_100"}, inplace=True)

df["bmi"] = (df["bmi"] > 21).astype(int)
df.rename(columns={"bmi":"bmi_above_21"}, inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
sns.heatmap(df.corr())
plt.savefig("heatmap.png")
plt.show()
```

```python
df.head()
```

### Algotithms available
- chow-liu
- greedy
- exact
- exact-dp

```python
import numpy as np
from pomegranate import *
model = BayesianNetwork.from_samples(df.to_numpy(), state_names=df.columns.values, algorithm='chow-liu')
model.plot()
plt.savefig("model-chow.png")
plt.show()
```

```python
import numpy as np
from pomegranate import *
model = BayesianNetwork.from_samples(df.to_numpy(), state_names=df.columns.values, algorithm='greedy')
model.plot()
plt.savefig("model-greedy.png")
plt.show()
```

```python
import numpy as np
from pomegranate import *
model = BayesianNetwork.from_samples(df.to_numpy(), state_names=df.columns.values, algorithm='exact')
model.plot()
plt.savefig("model-exact.png")
plt.show()
```

```python
import numpy as np
from pomegranate import *
model = BayesianNetwork.from_samples(df.to_numpy(), state_names=df.columns.values, algorithm='exact-dp')
model.plot()
plt.savefig("model-exact-dp.png")
plt.show()
```
